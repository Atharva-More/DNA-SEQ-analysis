```python
def find_coding_sequences(sequence):
    start_codon = "ATG"
    stop_codons = ["TAA", "TAG", "TGA"]
    
    coding_sequences = []
    start_index = sequence.find(start_codon)
    
    while start_index != -1:
        # Loop through the sequence in triplets
        for i in range(start_index, len(sequence), 3):
            codon = sequence[i:i+3]
            if codon in stop_codons:
                 
                    # Append the coding sequence to the list
                coding_sequences.append(sequence[start_index:i+3])
                # Move to the next start codon after the current stop codon
                start_index = sequence.find(start_codon, i+3)
                break
        else:
            # If no stop codon is found, exit the loop
            start_index = -1
    
    return coding_sequences
find_coding_sequences(sequence)
```

```python
def find_motifs(sequence, motif):
    print(f"\nSearching for motif: {motif}")
    matches = [match.start() for match in re.finditer(motif,sequence)]
    if matches:
        print(f"Motif '{motif}' found at positions: {matches}")
    else:
        print(f"Motif '{motif}' not found")
motif = "TATAA"     
find_motifs(sequence, motif)
```
